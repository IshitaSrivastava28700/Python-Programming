**1.** **Find a factorial of given number.**

```python
# Program to find factorial of a number using loops
num = int(input("Enter a number: "))
fact = 1
for i in range(1, num + 1):   # loop from 1 to num
    fact *= i                 # multiply fact by i
print("Factorial:", fact)
```

**2.**	**Find whether the given number is Armstrong number.**

```python
# Armstrong number: sum of cubes of digits = number itself
num = int(input("Enter a number: "))
temp = num
sum_val = 0
while temp > 0:
    digit = temp % 10         # extract last digit
    sum_val += digit ** 3     # cube and add
    temp //= 10               # remove last digit
print("Armstrong Number" if sum_val == num else "Not Armstrong")
```

**3.**	**Print Fibonacci series up to given term.**

```python
# Fibonacci series: 0,1,1,2,3,5,...
n = int(input("Enter number of terms: "))
a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b           # update values
print()
```

**4.**	**Write a program to find if given number is prime number or not.**

```python
# Prime number: divisible only by 1 and itself
num = int(input("Enter a number: "))
is_prime = True
if num < 2:
    is_prime = False
else:
    for i in range(2, int(num**0.5) + 1):  # check divisors up to sqrt(num)
        if num % i == 0:
            is_prime = False
            break
print("Prime" if is_prime else "Not Prime")
```

**5.**	**Check whether given number is palindrome or not.**

```python
# Palindrome: number/string same forwards and backwards
num = input("Enter a number: ")
print("Palindrome" if num == num[::-1] else "Not Palindrome")
```

**6.**  **Write a program to print sum of digits.**

```python
# Sum of digits of a number
num = int(input("Enter a number: "))
sum_digits = 0
while num > 0:
    sum_digits += num % 10    # add last digit
    num //= 10                # remove last digit
print("Sum of digits:", sum_digits)
```

**7.**	**Count and print all numbers divisible by 5 or 7 between 1 to 100.**

```python
# Program to print and count numbers divisible by 5 or 7 between 1 and 100

# Initialize a counter to keep track of how many numbers are found
count = 0

print("Numbers divisible by 5 or 7 between 1 and 100:")

# Loop through numbers from 1 to 100
for num in range(1, 101):
    # Check if the number is divisible by 5 or 7
    if num % 5 == 0 or num % 7 == 0:
        print(num, end=" ")   # Print the number on the same line
        count += 1            # Increase the counter by 1

# Print the total count at the end
print(f"\nTotal count: {count}")
```

**9.**	**Print the table for a given number:**
             **5 * 1 = 5**
            ......**5 * 2 = 10………..**


```python
# Program to print the multiplication table for a given number

# Take input from the user
num = int(input("Enter the number for its multiplication table: "))

# Print the table from 1 to 10
print(f"Multiplication table for {num}:")
for i in range(1, 11):  # Loop from 1 to 10
    result = num * i    # Multiply the number with i
    print(f"{num} * {i} = {result}")  # Print in table format
```

**10.**  **Write a program to print the following pattern**

123454321

1234 *4321

123  * * 321

12   * * *  21

1    * * * *   1


```python
# The number of rows for the pattern
rows = 5

# Outer loop to iterate through each row
for i in range(1, rows + 1):
    # Inner loop to print the ascending numbers
    # The range goes from 1 up to (rows - i + 1)
    for j in range(1, rows - i + 2):
        print(j, end="")

    # Inner loop to print the stars
    # The number of stars is 2*(i-1)
    for k in range(1, 2 * (i - 1) + 1):
        print("*", end="")

    # Inner loop to print the descending numbers
    # The range goes from (rows - i + 1) down to 1
    # The first row doesn't print this part as the range is invalid (5 down to 1 in this logic)
    # A conditional check is used to handle the first row's specific number sequence
    if i == 1:
        for l in range(rows - i, 0, -1):
            print(l, end="")
    else:
        for l in range(rows - i + 1, 0, -1):
            print(l, end="")

    # Move to the next line after the row is printed
    print()
```

**11.**	 **Write a program to print the sum of the following series
1+ ½ + 1/3 + ¼ +….+1/n**



```python
# Program to calculate the sum of the series:
# 1 + 1/2 + 1/3 + 1/4 + ... + 1/n

# Take input from the user for n
n = int(input("Enter the value of n: "))

# Initialize sum to 0
series_sum = 0

# Loop from 1 to n
for i in range(1, n + 1):
    series_sum += 1 / i   # Add the reciprocal of i to the sum

# Print the result
print(f"The sum of the series up to 1/{n} is: {series_sum}")
```

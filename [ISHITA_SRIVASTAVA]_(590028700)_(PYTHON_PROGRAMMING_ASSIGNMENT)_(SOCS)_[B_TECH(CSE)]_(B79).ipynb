**Difference between scripting and interactive modes in IDLE:-**



In Python’s IDLE, we can work in two modes: interactive mode and script mode. In interactive mode, we type commands directly into the Python shell, and each command is executed immediately after pressing Enter. This is useful for quick testing, calculations, or learning small pieces of code because we get instant results line by line. In contrast, script mode allows us to write a complete program in a separate file with a .py extension using the IDLE editor. After writing the program, we run the entire file at once, and the output is displayed in the shell. Script mode is better for writing longer programs, saving them for future use, and running them as a whole. In short, interactive mode is for immediate execution of single commands, while script mode is for writing, saving, and executing full programs.

**4.** **Take different data types and print values using print function.**

```python
# Demonstrating printing of different data types in Python

# Different data types
integer_val = 42                # int
float_val = 3.14159              # float
string_val = "Hello, Python!"    # string
boolean_val = True               # bool
list_val = [1, 2, 3, 4]          # list
tuple_val = ('a', 'b', 'c')      # tuple
dict_val = {"name": "Alice", "age": 25}  # dictionary
none_val = None                  # NoneType

# Printing each variable separately
print("Integer:", integer_val)
print("Float:", float_val)
print("String:", string_val)
print("Boolean:", boolean_val)
print("List:", list_val)
print("Tuple:", tuple_val)
print("Dictionary:", dict_val)
print("NoneType:", none_val)

# Printing multiple values in one statement (comma-separated)
print("\nMultiple values together:", integer_val, float_val, string_val, boolean_val)

# Using f-string for formatted output
print(f"\nFormatted Output -> Integer: {integer_val}, Float: {float_val}, String: '{string_val}'")

# Using str.format() method
print("\nUsing format(): Integer: {}, Float: {}, String: '{}'".format(integer_val, float_val, string_val))

# Printing with custom separator and end
print("\nCustom separator and end example:")
print(integer_val, float_val, string_val, sep=" | ", end=" <-- End of Line\n")
```

**8.** **Write a program to find area of triangle when length of sides are given.**

```python
import math

# Input lengths of the sides
a = float(input("Enter length of side a: "))
b = float(input("Enter length of side b: "))
c = float(input("Enter length of side c: "))

# Calculate semi-perimeter
s = (a + b + c) / 2

# Calculate area using Heron's formula
area = math.sqrt(s * (s - a) * (s - b) * (s - c))

# Print result
print("The area of the triangle is:", area)
```

**9.** **Write a program to convert given seconds into hours, minutes and remaining seconds.**

```python
# Program to convert seconds into hours, minutes and seconds

# Input total seconds
total_seconds = int(input("Enter total seconds: "))

# Calculate hours, minutes, and remaining seconds
hours = total_seconds // 3600
minutes = (total_seconds % 3600) // 60
seconds = total_seconds % 60

# Print result
print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", seconds)
```

**11.** **Write a program to find sum of first n natural numbers.**

```python
# Program to find sum of first n natural numbers

try:
    n = int(input("Enter a positive integer: "))
    if n < 1:
        print("Please enter a number greater than 0.")
    else:
        total = n * (n + 1) // 2  # Formula-based calculation
        print(f"Sum of first {n} natural numbers is: {total}")
except ValueError:
    print("Invalid input! Please enter an integer.")

```

**12.** **Write a program to print truth table for bitwise operators (&, | and ^ operators)**


```python
# Truth table for bitwise &, |, and ^ operators
print("A B | A&B A|B A^B")
print("-"*20)
for A in (0, 1):
    for B in (0, 1):
        print(f"{A} {B} |  {A & B}   {A | B}   {A ^ B}")
```

**13.** **Write a program to find left shift and right shift values of a given number.**

```python
# Program to find left shift and right shift values

try:
    num = int(input("Enter a number: "))
    shift = int(input("Enter shift amount: "))

    # Left shift (<<) multiplies by 2^shift
    left_shift = num << shift
    # Right shift (>>) divides by 2^shift (floor division)
    right_shift = num >> shift

    print(f"{num} << {shift} = {left_shift}")
    print(f"{num} >> {shift} = {right_shift}")

except ValueError:
    print("Please enter valid integers.")
```

**14.** **Using membership operator find whether a given number is in sequence (10,20,56,78,89)**

```python
# Program to check if a number is in a given sequence using membership operator

# Define the sequence (tuple)
sequence = (10, 20, 56, 78, 89)

try:
    # Take user input
    num = int(input("Enter a number to check: "))

    # Use membership operator 'in'
    if num in sequence:
        print(f"{num} is present in the sequence {sequence}.")
    else:
        print(f"{num} is NOT present in the sequence {sequence}.")

except ValueError:
    print("Invalid input! Please enter an integer.")
```

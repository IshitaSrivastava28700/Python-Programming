**1. Check whether the given number is divisible by 3 and 5 both.**

```python
# Program to check divisibility by 3 and 5

# Taking input from user
num = int(input("Enter a number: "))

# Checking divisibility
if num % 3 == 0 and num % 5 == 0:
    print(num, "is divisible by both 3 and 5")
else:
    print(num, "is NOT divisible by both 3 and 5")
```

**2. Check whether a given number is multiple of five or not.**

```python
# Program to check if a number is a multiple of 5

# Taking input from user
num = int(input("Enter a number: "))

# Checking divisibility by 5
if num % 5 == 0:
    print(num, "is a multiple of 5")
else:
    print(num, "is NOT a multiple of 5")
```

**3. Find the greatest among the two numbers. If numbers are equal than print “numbers are equal”.**

```python
# Program to find the greatest among two numbers

# Taking input from user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Comparing numbers
if num1 > num2:
    print(num1, "is greater than", num2)
elif num2 > num1:
    print(num2, "is greater than", num1)
else:
    print("Numbers are equal")
```

**4. Find the greatest among three numbers assuming no two values are same.**

```python
# Program to find the greatest among three numbers (no two values are same)

# Taking input from user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

# Comparing numbers
if num1 > num2 and num1 > num3:
    print(num1, "is the greatest")
elif num2 > num1 and num2 > num3:
    print(num2, "is the greatest")
else:
    print(num3, "is the greatest")
```

**5. Check whether the quadratic equation has real roots or imaginary roots. Display the roots.**

```python
# Program to check nature of roots of a quadratic equation
# Equation: ax^2 + bx + c = 0

import math

# Taking coefficients as input
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

# Calculating discriminant
D = b**2 - 4*a*c

# Checking nature of roots
if D > 0:
    root1 = (-b + math.sqrt(D)) / (2*a)
    root2 = (-b - math.sqrt(D)) / (2*a)
    print("Roots are real and distinct.")
    print("Root 1 =", root1, ", Root 2 =", root2)
elif D == 0:
    root = -b / (2*a)
    print("Roots are real and equal.")
    print("Root =", root)
else:
    realPart = -b / (2*a)
    imagPart = math.sqrt(-D) / (2*a)
    print("Roots are imaginary (complex).")
    print("Root 1 =", realPart, "+", imagPart, "i")
    print("Root 2 =", realPart, "-", imagPart, "i")
```

**6. Find whether a given year is a leap year or not.**


```python
# Program to check if a year is a leap year

year = int(input("Enter a year: "))

# Leap year conditions:
# 1. Divisible by 400 → Leap year
# 2. Divisible by 4 but not by 100 → Leap year
# Otherwise → Not a leap year

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, "is a leap year")
else:
    print(year, "is NOT a leap year")
```

**7. Write a program which takes any date as input and display next date of the calendar
e.g.
I/P: day=20 month=9 year=2005  
O/P: day=21 month=9 year 2005**

```python
# Program to display the next date of the calendar

import datetime

# Taking input from user
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

# Create a date object
current_date = datetime.date(year, month, day)

# Add one day
next_date = current_date + datetime.timedelta(days=1)

# Display result
print("Next date is: day =", next_date.day,
      "month =", next_date.month,
      "year =", next_date.year)
```

**8. Print the grade sheet of a student for the given range of cgpa. Scan marks of five subjects and
calculate the percentage.
CGPA=percentage/10
CGPA range:
0 to 3.4 -> F
3.5 to 5.0->C+
5.1 to 6->B
6.1 to 7-> B+
7.1 to 8-> A
8.1 to 9->A+
9.1 to 10-> O (Outstanding)
Sample Gradesheet
Name: Rohit Sharma
Roll Number: R17234512   
Sem: 1      
Subject name: Marks
PDS:   
70
Python:  80
Chemistry:  
English:  60
Physics:  50
90
Percentage: 70%
CGPA:7.0
Grade:   
SAPID: 50005673
Course: B.Tech. CSE AI&ML**

```python
# Program to print grade sheet of a student

# Student details
name = "Rohit Sharma"
roll_number = "R17234512"
sem = 1
sapid = "50005673"
course = "B.Tech. CSE AI&ML"

# Marks of five subjects
marks = {
    "PDS": 70,
    "Python": 80,
    "Chemistry": 90,
    "English": 60,
    "Physics": 50
}

# Calculate percentage
total_marks = sum(marks.values())
percentage = total_marks / len(marks)

# Calculate CGPA
cgpa = percentage / 10

# Determine grade based on CGPA range
if 0 <= cgpa <= 3.4:
    grade = "F"
elif 3.5 <= cgpa <= 5.0:
    grade = "C+"
elif 5.1 <= cgpa <= 6.0:
    grade = "B"
elif 6.1 <= cgpa <= 7.0:
    grade = "B+"
elif 7.1 <= cgpa <= 8.0:
    grade = "A"
elif 8.1 <= cgpa <= 9.0:
    grade = "A+"
elif 9.1 <= cgpa <= 10.0:
    grade = "O (Outstanding)"
else:
    grade = "Invalid CGPA"

# Print grade sheet in the same format as question
print("------ Grade Sheet ------")
print("Name:", name)
print("Roll Number:", roll_number)
print("Sem:", sem)
print("Subject name: Marks")
for subject, mark in marks.items():
    print(subject + ":", mark, end="  ")
print("\nPercentage:", str(int(percentage)) + "%")
print("CGPA:", cgpa)
print("Grade:", grade)
print("SAPID:", sapid)
print("Course:", course)
```

**1.** **Write a program to count and display the number of capital letters in a given string.**

```python
# Program to count capital letters in a string

text = input("Enter a string: ")
count = 0

for ch in text:
    if ch.isupper():   # Check if character is uppercase
        count += 1

print("Number of capital letters:", count)
```

 **2.**  **Count total number of vowels in a given string.**

```python
# Program to count vowels in a string

text = input("Enter a string: ")
vowels = "AEIOUaeiou"
count = 0

for ch in text:
    if ch in vowels:   # Check if character is a vowel
        count += 1

print("Number of vowels:", count)
```

**5.** **Given a string containing both upper and lower case alphabets. Write a Python program
to count the number of occurrences of each alphabet (case insensitive) and display the
same.**

**Sample Input**

**ABaBCbGc**

**Sample Output**

**2A**

**3B**

**2C**

**1G**


```python
# Program to count occurrences of each alphabet (case-insensitive)

string = input("Enter a string: ")

# Convert to uppercase for uniform counting
string = string.upper()

# Dictionary to store counts
counts = {}

for ch in string:
    if ch.isalpha():   # check if character is an alphabet
        counts[ch] = counts.get(ch, 0) + 1

# Display results
for letter, count in counts.items():
    print(f"{count}{letter}")
```

**6.** **Program to count number of unique words in a given sentence using sets**.

```python
# Program to count number of unique words in a sentence using sets

sentence = input("Enter a sentence: ")

# Split sentence into words
words = sentence.split()

# Convert list of words into a set (removes duplicates)
unique_words = set(words)

# Count unique words
print("Number of unique words:", len(unique_words))

# Display the unique words
print("Unique words are:", unique_words)
```

**7.** **Create 2 sets s1 and s2 of n fruits each by taking input from user and find:**

**a) Fruits which are in both sets s1 and s2**

**b) Fruits only in s1 but not in s2**

**c) Count of all fruits from s1 and s2**

```python
# Program to work with two sets of fruits

# Input for first set
n1 = int(input("Enter number of fruits in set s1: "))
s1 = set()
for i in range(n1):
    fruit = input(f"Enter fruit {i+1} for set s1: ")
    s1.add(fruit)

# Input for second set
n2 = int(input("Enter number of fruits in set s2: "))
s2 = set()
for i in range(n2):
    fruit = input(f"Enter fruit {i+1} for set s2: ")
    s2.add(fruit)

# a) Fruits in both sets
print("Fruits in both sets:", s1 & s2)

# b) Fruits only in s1 but not in s2
print("Fruits only in s1:", s1 - s2)

# c) Count of all fruits from s1 and s2
all_fruits = s1 | s2
print("Total number of unique fruits in both sets:", len(all_fruits))
```

**8.** **Take two sets and apply various set operations on them :**

**S1 = {Red ,yellow, orange , blue }**

**S2 = {violet, blue , purple}**

```python
# Define the sets
S1 = {"Red", "Yellow", "Orange", "Blue"}
S2 = {"Violet", "Blue", "Purple"}

# Union
print("Union (S1 ∪ S2):", S1.union(S2))

# Intersection
print("Intersection (S1 ∩ S2):", S1.intersection(S2))

# Difference (S1 - S2)
print("Difference (S1 - S2):", S1.difference(S2))

# Difference (S2 - S1)
print("Difference (S2 - S1):", S2.difference(S1))

# Symmetric Difference
print("Symmetric Difference (S1 ⊕ S2):", S1.symmetric_difference(S2))
```
